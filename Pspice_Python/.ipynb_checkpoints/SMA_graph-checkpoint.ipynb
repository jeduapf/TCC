```python
import cmath
import numpy as np
import matplotlib.pyplot as plt

points = 1000
w = np.linspace(0.00001,np.pi,points)

def window_filter(L,vec):
    return [(1/L)*(1-cmath.exp(complex(0,-w*L)))/(1-cmath.exp(complex(0,-w)))for w in vec]

plt.figure(figsize=(16,5))
for L in [1,2,3,4]:
    plt.plot(w[0:500],20*np.log(np.abs(window_filter(L,w))[0:500]), label='SMA'+str(L))
plt.axhline(y = -3, color = 'k', linestyle = '--', label = '-3dB')
plt.legend()
plt.ylabel('|H(jw)|[dB]')
plt.xlabel('w[rad]')
plt.savefig("SMA.png", dpi=500)
```
