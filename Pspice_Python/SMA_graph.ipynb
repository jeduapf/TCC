```python
import cmath
import numpy as np
import matplotlib.pyplot as plt

points = 10**5
w = np.linspace(0.00001,np.pi,points)

def window_filter(L,vec):
    return [(1/L)*(1-cmath.exp(complex(0,-w*L)))/(1-cmath.exp(complex(0,-w)))for w in vec]

P = 6*10**3
plt.figure(figsize=(16,5))
for L in [10,20,30]:
    plt.plot(w[0:P],20*np.log(np.abs(window_filter(L,w))[0:P]), label='SMA'+str(L))
plt.axhline(y = -3, color = 'k', linestyle = '--', label = '-3dB')
plt.axvline(x = 2*np.pi*10/10**3, color = 'r', linestyle = '--')
plt.axvline(x = 2*np.pi*15/10**3, color = 'r', linestyle = '--')
plt.axvline(x = 2*np.pi*30/10**3, color = 'r', linestyle = '--')
plt.legend()
plt.ylabel('|H(jw)|[dB]')
plt.xlabel('w[rad]')
plt.savefig("SMA.png", dpi=500, bbox_inches='tight')
```

```python
# Freq de corte de 60 Hz => 7371 pontos se fa = 1MHz
points = 10**6
w = np.linspace(0.00001,np.pi,points)

plt.figure(figsize=(16,5))
K = 1000
plt.plot(w[0:K],20*np.log(np.abs(window_filter(7371,w))[0:K]), label='SMA'+str(7371))
plt.axhline(y = -3, color = 'k', linestyle = '--', label = '-3dB')
plt.axvline(x = 2*np.pi*40/10**6, color = 'r', linestyle = '--', label = '60Hz')
plt.legend()
plt.ylabel('|H(jw)|[dB]')
plt.xlabel('w[rad]')
plt.savefig("Filtro_60Hz.png", dpi=500, bbox_inches='tight')
```
